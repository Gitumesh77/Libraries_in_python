Joining the Arrays

```python
# vertical, horizontal and column stack
```

```python
#vertical
import numpy as np
n1 = np.array([1,2,3,4])
n2 = np.array([2,3,4,5])
np.vstack((n1,n2))
```

```python
# horizontal stack
n1 = np.array([1,2,3,4])
n2 = np.array([2,3,4,5])
np.hstack((n1,n2))
```

```python
# column stack
n1 = np.array([1,2,3,4])
n2 = np.array([2,3,4,5])
np.column_stack((n1,n2))
```

Intersection and difference

```python
# intersection
n1 = np.array([1,2,3,4])
n2 = np.array([2,3,5,6])
np.intersect1d(n1,n2)
```

```python
# difference 
n1 = np.array([1,2,3,4])
n2 = np.array([2,3,5,6])
np.setdiff1d(n1,n2)
```

```python
# difference 
n1 = np.array([1,2,3,4])
n2 = np.array([2,3,5,6])
np.setdiff1d(n2,n1)
```

BAR PLOT

```python
# create a dictionary to load the details
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

student = ({'Bob':87, 'Julia':45, 'Anna': 100, 'Matt':67})
```

```python
names = list(student.keys())
print(names)
```

```python
values = list(student.values())
values
```

```python
plt.bar(names,values)
plt.show()
# by analysing the chart we found that Anna has scored highest marks
```

```python
# to assign the labels for x and y axis
plt.bar(names,values)
plt.title("Bar chart")
plt.xlabel("names")
plt.ylabel("marks")
plt.grid(True)
plt.show()
```

Scatter plot

```python
x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]
plt.scatter(x,a)
plt.show()
```

```python
# changing the colour, sign and size of the dots
x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]
plt.scatter(x,a, marker = "*", c= "Red", s= 100)
plt.show()
```

```python
# adding 2 markers in the same plot 
x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]
b = [5,2,1,7,5,6,1,0,4]
plt.scatter(x,a, marker = "*", c= "Red", s= 100)
plt.scatter(x,b, marker = ".", c= "Blue", s= 80)
plt.show()
```

```python
# adding subplots (creating 2 plots in a single page)
x = [10,20,30,40,50,60,70,80,90]
a = [8,1,7,2,0,3,7,3,2]
b = [5,2,1,7,5,6,1,0,4]

plt.subplot(1,2,1)
plt.scatter(x,a, marker = "*", c= "Red", s= 100)

plt.subplot(1,2,2)
plt.scatter(x,b, marker = ".", c= "Blue", s= 80)
plt.show()
```

Seaborn Plot

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(x='timepoint', y = 'signal', data=fmri)
plt.show()
# download the file by clicking link (https://github.com/mwaskom/seaborn-data/blob/master/fmri.csv)
```

```python
# grouping data with hue
sns.lineplot(x='timepoint', y = 'signal', data=fmri, hue= 'event')
plt.show()
```

```python
sns.lineplot(x='timepoint', y = 'signal', data=fmri, hue= 'event', style= 'event')
plt.show()
```

```python
# setting the markers
sns.lineplot(x='timepoint', y = 'signal', data=fmri, hue= 'event', style= 'event', markers = True)
plt.show()
```
